**Ordinary Differential Equation - Initial Value Problems**

**22.1 ODE initial Value Problem statement**

Differential equation: It's the relationship between function f(x) and independent variable x, and the derivatives of x.

Ordinary differential equation: Differnetial equations where the independent variable x and its derivatives in one dimension

Example:

Suppose F(x,f(x), [df(x)/dx], [d^2f(x)/dx^2], [d^3f(x)/dx^3], ... , d^n-1f(x)/dx^n-1) = d^nf(x)/dx^n

Then this equation is n-th order ordinary differential equation and n is the order of the differential equation

Partial differential equation: General form differential equation where x is a vector that contains multiple independent variables x_1, x_2, ... , x_m. The order of partial derivatives depends on the order of the conbination of the vectors

General solution of a differential equation is a function g(x) that satisfies the differential equation. A particule solution is a function p(x) that satisfies the differential equation given that there n explicitly known values of the solution at certain point.

**22.2 Reduction of order**

It's a method to solve a higher order differential equation given that the solutions for previous order diffferential equation is know.

Example: 

Suppose the n-th order ordinary differential equation:

f^n(t) = F(t, f(t), f^1(t), f^2(t), f^3(t), ... , f^(n-1)(t))

Define the Vector S(t) that contains the ordinary differential equation from order 1 to order (n-1)

Then ,take the derivatives of S(t) to form a new vector. Then the last row of the new vector contains the the equation f^n(t). Which is the solution for n-th order differential equation.


**22.3 Ruler Method**

It's a method to approximate the solution of a ordinary differential equation by making n equivalent discrete intervals of the differential equation. First is to solve differential equation based on the first interval, then based the solution over the previous interval to solve for the equation over the next interval. Repeat the process until the solution on (n-1) interval is solved

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: np.cos(2*t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 9))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
