**Elements of Linear Algebra**

1: Linear spaces
Let V = R^n
(1) Linear combinations: A linear combination is a vector as the output from a subset by multiplying each vector in that subset by a constant and add them together. The linear subspace is the space that contains all the possiable linear combinations. 

Define the subset U as the linear subspace of V. Then U has three properties:

(a) for any vectors u_n and u_(n+1), the addition of them is still in the linear subspace; 

For example: let u_n and u_(n+1) be 1 by 3 vectors that u_n= [1 2 3] and u_(n+1)=[4 5 6], then the addtion of them u_n+u_(n+1)=[1 2 3]+[4 5 6]=[5 7 9] is still in the linear subspace U


(b) multiply any vector in the linear subspace by any constant that is a real number, the product of them is still in the linear subspace.

For example: let u_n be a 1 by 4 vector that u_n=[3 3 3 3], and let b be the real number 3, then 3* u_n=[9 9 9 9] is still in the linear subspace U


(c) The vector 0 is always in a linear subspce.


2: Spanning set of an Linear subspace:

(a)Suppose there are number of n vectors in V: {w_1, w_2, w_3, ... , w_n}, then the span(w_1, ... ,w_n) is the set of all possiable linear combinations of w_j,j=1,2,3,...,n.

(b)A span itself is s linear subspace.

3: Column space

Let A be a 3 by 3 matrix with columns [1 2 3], [3 2 1], [4 5 6]. Then the column space, denoted as col(A), is the span of the columns of A. In other words, the space that contains all the possiable linear combinations of the three columns is the column space of A

For example: 
[1 2 3] could be in the column space since 1*[1 2 3]+ 0*[3 2 1]+ 0*[4 5 6] equals to [1 2 3].

[4 4 4] is also in the column space since 1*[1 2 3]+1*[3 2 1]+0*[4 5 6]=[4 4 4] 

4: Linear independence and dimension

(1) Linear independence: A set of vectors are said to be linearly independent if none of them can't be weitten as a linear combina of others. A set of vectors are said to be linearly dependent if they are not independent.

A set of vectors is linearly independent if and only if the linear combination of them is the vector 0 when the scalar of each vector is the real number 0.

For example:
(a) vectors [1 1 1] [2 2 2] [3 3 3] are linearly dependent because [1 1 1]+[2 2 2]=[3 3 3]. Therefore, vector [3 3 3] is a linear combination of [1 1 1] and [2 2 2].

(b) vectors [1 0 0] [0 2 0] [0 0 3] are linearly independent because there is no linear combination of any two vectors in this set is equivalent to another vector. And to produce a 0 vector from those three vectors, the only option is to multiply each vector by 0 and add them up or minus them.


(2) Basis of a space: Suppose U is a linear subspace of V. A basis of U is a set of vectors that statisfies two conditions: (a) They are the span of U
(b) They are linearly independent.

For example:

two 2 by 1 vectors [1 0] and [0 1] can be the basis for a R^2 subspace since the linear combination of them can form any 1 by 2 vector and they are independent to each other.




5: Dimension theorem. Suppose U is a linear subspace of V. Then all bases of U have the same number elements(or same length), the number of elements that a basis has is the dimension of the linear subspace U, denoted as dim(U)

Example: Let vectors [1 0] and [0 1] form a basis for U, then U is 2 dimension or dim(U)=2

6: Orthogonality:

A list of vectors is said to be orthonormal if the product of each two different vectors in this set is 0 and the the inner product of each vector in this set is 1.

Example:

Let S be a set that contains the vectors: {[-1/sqrt(2),1/sqrt(2)], [1/sqrt(2), 1/sqrt(2)]}. 

Then the product of those two vector is (-1/sqrt(2)*1/sqrt(2))+(1/sqrt(2)*1/sqrt(2))=(-1/2)+(1/2)=0. 

The inner product of vector [-1/sqrt(2),1/sqrt(2)] is (-1/sqrt(2))^2+(1/sqrt(2))^2=(1/2)+(1/2)=1. 

The inner product of vector [1/sqrt(2),1/sqrt(2)] is (1/sqrt(2))^2+(1/sqrt(2))^2=(1/2)+(1/2)=1.

Since the product of those two vectors is 0 and the inner product of each vector is 1, therefore, this set of vectors is said to be orthonormal.
