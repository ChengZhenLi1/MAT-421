**Section 17.4: Lagrange polynomial interpolation**

Lagrange polynomial is a single polynomial that goes through all the data points in a data set.

Linear combination of lagrange basis polynomials is a way to construct the lagrange polynomial. Lagrange basis polynomials have the properties that P_i(x_j)=1 when i=j and  P_i(x_j)=0 when i doesn't equal to j.

Below are the code and graph to show those 2 properties

```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

# set up values for x and y 
x = [0, 1, 2]
y = [1, 3, 2]
P1_coeff = [1,-1.5,.5]
P2_coeff = [0, 2,-1]
P3_coeff = [0,-.5,.5]

# get the polynomial function
P1 = poly.Polynomial(P1_coeff)
P2 = poly.Polynomial(P2_coeff)
P3 = poly.Polynomial(P3_coeff)

x_new = np.arange(-1.0, 3.1, 0.1)

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, P1(x_new), 'b', label = 'P1')
plt.plot(x_new, P2(x_new), 'r', label = 'P2')
plt.plot(x_new, P3(x_new), 'g', label = 'P3')

plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')
plt.title('Lagrange Basis Polynomials')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()

# combine those three lagrange basis polynomials
L = P1 + 3*P2 + 2*P3

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, L(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

From the graphs above, the first graph shows that the values for each lagrange basis polynomials. 

For i=1, P_1(x_1)=1 ,  P_1(x_2)=0 ,  P_1(x_3)=0 

For i=2, P_2(x_2)=1  , P_2(x_1)=0 ,  P_3(x_3)=0

For i=3, P_3(x_3)=1  , P_3(x_1)=0  , P_3(x_2)=0

Which satisfying the property of lagrange basis polynomials

The sccond graph shows that a linear combination of lagrange basis polynomials that forms a lagrange polynomial that goes through the three points (0,1),(1,3),(2,2).



Another way to do Lagrange polynomial interpolation is to call the function 'lagrange' direcly. Below are the example of using lagrange function to do the lagrange polynomial interpolation

```python
from scipy.interpolate import lagrange

#set up the values for x and y 
x=[-3,6,5]
y=[4,6,5]
x_new = np.arange(-4.0, 7, 0.1)

#call lagrange to interpolate the data
f = lagrange(x, y)
fig = plt.figure(figsize = (10,8))
plt.plot(x_new, f(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Section 17.5 Newton's polynomial interpolation**

Nweton's polynomial  interpolation is a method that fits exactly for a set of data points. Its general form for (n-1) order Newton's polynomial that goes through n points is: 

f(x)=a_0+a_1(x-x_0)+a_2(x-x_0)(x-x_1)+...+a_n(x-x_0)(x-x_1)...(x-x_n)

To find the coefficients a_i, we use the method called divided differences, the formula is :

 f(x_k, x_(k-1), ... ,x_1, x_0)=(f(x_k,x_(k-1),...,x_2,x_1)-f(x_k,x_(k-1),...,x_1,x_0))/(x_k-x_0)

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

%matplotlib inline
```

```python
def divided_diff(x, y):
    '''
    function to calculate the divided
    differences table
    '''
    n = len(y)
    coef = np.zeros([n, n])
    # the first column is y
    coef[:,0] = y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
            
    return coef

def newton_poly(coef, x_data, x):
    '''
    evaluate the newton polynomial 
    at x
    '''
    n = len(x_data) - 1 
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p
    return p
```

```python
x = np.array([-5, -3, 0, 2])
y = np.array([-2, -7, -10, 7])
# get the divided difference coef
a_s = divided_diff(x, y)[0, :]

# evaluate on new data points
x_new = np.arange(-5, 2.1, .1)
y_new = newton_poly(a_s, x, x_new)

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo')
plt.plot(x_new, y_new)
```

From the graph above, it shows that the newton's polynomial goes through all the data points
