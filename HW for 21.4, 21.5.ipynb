**Simpson's rule**

It's a method to approximate the intergral of function f(x) over two subintervals. If there exists two consecutive subintervals [x_(i-1),x_i] and [x_i,x_(i+1)] for function f(x). By using simpson's rule, first fitting a quadractic polynomial that goes through the points (x_(i-1),f(x_(i-1))), (x_(i),f(x_(i))), (x_(i+1),f(x_(i+1))). Then integrating the quadratic euqation and using the value of that as the approximation for the value of intergal f(x) in the interval (x_(i-1),x_(i+1))

```python
import numpy as np

a = 3
b = np.pi
n = 13
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) \
            + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = 2 - I_simp

print(I_simp)
print(err_simp)
```

**Computing integrals in python**

trapz function to approximate integral of a function :

```python
import numpy as np
from scipy.integrate import trapz

# approximate the integral of cos(x) in the interval (0, pi+2) with 13 equally 
# spaced points
a = 0
b = np.pi+2
n = 13
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_trapz = trapz(f,x)
I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])

print(I_trapz)
print(I_trap)
```

cumtrapz function to approximate the cumulative integral of a function.

```python
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-poster')

x = np.arange(0.1, np.pi, 0.02)
F_exact = np.sin(x)
F_approx = cumtrapz(np.cos(x), x)

plt.figure(figsize = (10,13))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_0^{x} cos(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()
```

quad function to approximate the integral of a function

```python
from scipy.integrate import quad 
# compute the integral of cos(x) with the interval (2, pi)

I_quad, est_err_quad = \
          quad(np.cos, 2, np.pi)
print(I_quad)
err_quad = 2 - I_quad
print(est_err_quad, err_quad)
```
